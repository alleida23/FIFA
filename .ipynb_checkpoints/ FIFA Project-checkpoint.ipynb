### Dealing with NaN values

```python
# Looking for columns containing null values.
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
# Dropping 'Loan date end' and 'id' (not addit. info) , 'team&contract' (info repeated)
data = data.drop(['loan date end','id','team & contract'], axis=1)

null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
# Various columns '41' Nan value are the same rows --> drop them
data = data.dropna(subset=['joined','volleys','curve','agility','balance','jumping','vision','sliding tackle'])

#checking
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
# 'Unknown' --> club, position, a/w and d/w

# Replacing na values with'Unkown'
data["club"] = data["club"].fillna("Unknown", inplace = True)
data["position"] = data["position"].fillna("Unknown", inplace = True)
data["a/w"] = data["a/w"].fillna("Unknown", inplace = True)
data["d/w"] = data["d/w"].fillna("Unknown", inplace = True)


#checking results
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
# Composure --> fill with median (?)

#data['composure'] = data['composure'].fillna(value=mean_composure, inplace=True)
data['composure'] = data['composure'].fillna(data['composure'].mean())
#data['composure'] = data['composure'].fillna(data['composure'].median())

#checking results
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python

def clean_data(df):
    
    #Lowercase
    
    cols = []
    for c in df.columns:
        cols.append(c.lower())
    df.columns = cols
    
    #Drop Duplicates
    df = df.drop_duplicates()
    
    #Drop unnecesary columns
    df = df.drop(['id','age','growth','team & contract'], axis = 1)
    
    # Dropping NaN colum
    df = df.drop(['loan date end'], axis=1)
    
    # Dropping NaN rows
    df = df.dropna(subset=['joined','volleys','curve','agility','balance','jumping','vision','sliding tackle'])
    
    # Replacing Nans with mean value of the column
    df['composure'] = df['composure'].fillna(df['composure'].mean())
    
    # Delete Star symbols
    df['ir']=df['ir'].str.replace('★','')
    df['w/f']=df['w/f'].str.replace('★','')
    df['sm']=df['sm'].str.replace('★','')
    
    # Delete or replace characters
    df['height']=df['height'].str.replace("\'",".").str.replace('\"','')
    df['weight']=df['w/f'].str.strip('lbs')
    df['value'] =df['value'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
    df['wage'] = df['wage'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
    df['release clause'] = df['release clause'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
    
    # Datetime
    
    #from datetime import datetime
    df['joined'] = pd.to_datetime(df['joined'])
    df['year'] = df['joined'].dt.year
    #df['joined'] = datetime.strftime("%Y")
    
    # Split contract in 2 columns: start / end
    #split_df = df['contract'].str.split('(\~)', expand=True)
    #df['contract from'] = pd.to_numeric(split_df[1])
    #df['contract until'] = split_df[0].replace('[\~]', '', regex=True)
    
    return df
    
    
    
  
```

```python
data_test = pd.read_csv('fifa21_train.csv')
clean_data(data_test)
```
